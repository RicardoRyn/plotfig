## 全脑

### 快速出图

脑区图使用了`surfplot`的API，结合图集文件，成对对某块脑区值的绘制。

目前支持图集包括：

1. 人 Glasser图集[^1]
1. 人 BNA图集[^2]
1. 猕猴 CHARM 5-level图集[^3]
1. 猕猴 CHARM 6-level图集[^3]
1. 猕猴 BNA图集[^4]
1. 猕猴 D99图集[^5]
1. 黑猩猩 BNA图集[^6]

例如绘制人类Glasser图集中左脑V1区域值为1，右脑MT区域值为1.5。
